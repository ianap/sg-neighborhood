## Reading data: 

```python
with open('buildings.json') as json_data:
    singapore_data = json.load(json_data)
```

```python
neighborhoods_data = singapore_data[0]
```

### Tranform the data into a pandas dataframe

```python
# define the dataframe columns
column_names = ['Postcode', 'Neighborhood', 'Latitude', 'Longitude'] 

# instantiate the dataframe
neighborhoods = pd.DataFrame(columns=column_names)
```

```python
for data in singapore_data:
    post = data['POSTAL']
    neighborhood_name = data['BUILDING']
        
    neighborhood_lat = data['LATITUDE']
    neighborhood_lon = data['LONGITUDE']
    
    neighborhoods = neighborhoods.append({'Postcode': post,
                                          'Neighborhood': neighborhood_name,
                                          'Latitude': neighborhood_lat,
                                          'Longitude': neighborhood_lon}, ignore_index=True)
```

## Saving data into csv file for the next analysis.

```python
neighborhoods.head()
```

```python
neighborhoods.to_csv('singapore_geo.csv', encoding='utf-8')
```
